```python
import basis
import numpy as np
import scipy as sp
import scipy.linalg
import plotly.graph_objects as go
```

```python
np.random.seed(basis.SEED)
y = np.arange(-1, 2, 0.05)
yrand = y + np.random.random(y.shape)
xdata = np.ones((y.shape[0], 4))
t1 = np.arange(1, 5.2, 0.07)
t2 = np.arange(-7, -1, 0.1)
xdata[:, 2] = np.sin(2*np.pi*t1)
xdata[:, 3] = np.cos(2*np.pi*t1)
xdata_rand = xdata + (np.random.random(xdata.shape) - 0.5) / 10
xdata_rand[:, 0] = xdata[:, 0]
xdata_rand[:, 1] = t1[:]
xdata_pinv = sp.linalg.pinv(xdata_rand)

theta_hat = xdata_pinv @ yrand
y_hat = xdata_rand @ theta_hat
y_hat_clean = xdata @ theta_hat

fig = go.Figure(
    data=[
        go.Scatter3d(
            x=xdata_rand[:, 2],
            y=xdata_rand[:, 3],
            z=yrand,
            mode="markers",
            marker=dict(
                size=[2],
                sizemode="diameter",
            ),
            name="Sample data",
        ), go.Scatter3d(
            x=xdata_rand[:, 2],
            y=xdata_rand[:, 3],
            z=y_hat,
            mode="lines",
            name="Standard Least Squares Best Fit",
        ),
    ],
)
fig.show()
```
